```python
import nibabel as nib
import os
import numpy as np
import matplotlib.pyplot as plt
import math
from tqdm.notebook import tqdm
import pickle
from sklearn.preprocessing import StandardScaler
import cv2
```

```python
level = 36
```

```python
dataset = []
image = nib.load('./Dataset/CN/2006-08-14_06_16_40.0/I20470/ADNI_036_S_0672_PET_30_MIN_3D_FDG_4i_16s__br_raw_20060814124938578_1_S17804_I20470.nii')
dataset.append(np.array(image.get_fdata())[:,:,:,0]/np.array(image.get_fdata())[:,:,:,0].max())
dataset = np.array(dataset)
subject = dataset[0]

#Getting Thresholds
ret,thresh_low = cv2.threshold(subject,0.35,255,cv2.THRESH_BINARY)
ret,thresh_med = cv2.threshold(subject,0.55,255,cv2.THRESH_BINARY)
ret,thresh_high = cv2.threshold(subject,0.7,255,cv2.THRESH_BINARY)
plt.imshow(subject[:,:,level],cmap='jet')
plt.title("Cognitively Normal")
plt.show()
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(15,15))

#With JET Map
ax1.imshow(cv2.bitwise_and(thresh_low[:,:,level],subject[:,:,level]),cmap='jet')
ax2.imshow(cv2.bitwise_and(thresh_med[:,:,level],subject[:,:,level]),cmap='jet')
ax3.imshow(cv2.bitwise_and(thresh_high[:,:,level],subject[:,:,level]),cmap='jet')
plt.show()
#Without JET Map
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(15,15))
ax1.imshow(thresh_low[:,:,level],cmap='jet')
ax2.imshow(thresh_med[:,:,level],cmap='jet')
ax3.imshow(thresh_high[:,:,level],cmap='jet')
plt.show()
```
